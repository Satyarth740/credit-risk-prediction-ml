```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train_df = pd.read_csv("../Data/train_clean.csv")
```

```python
sns.countplot(x="Risk", data=train_df)
plt.title("Credit Risk Distribution")
plt.show()
```

```python
sns.boxplot(x="Risk", y="ApplicantIncome", data=train_df)
plt.show()
```

```python
sns.boxplot(x="Risk", y="LoanAmount", data=train_df)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(train_df.select_dtypes("number").corr(), annot=True, cmap="coolwarm")
plt.show()
```
